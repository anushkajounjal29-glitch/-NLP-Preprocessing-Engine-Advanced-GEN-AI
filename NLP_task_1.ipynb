Task 1: Conceptual Understanding

1. Difference between "Love" and "love" in NLP
In NLP, text is case-sensitive by default.
"Love" and "love" are treated as different tokens.
This can increase vocabulary size unnecessarily.
That’s why we convert text to lowercase.


2. What happens if stopwords are not removed?
Model gets extra noise (e.g., "is", "the", "and")
Increases processing time and memory
May reduce model accuracy in some cases

3. When removing stopwords can be harmful
Sentiment Analysis
Example: "I am not happy" → removing "not" changes meaning
Question Answering Systems
Example: "What is AI?" → removing "what" breaks intent

4. Stemming vs Lemmatization
Feature	Stemming	Lemmatization
Method	Cuts words	Uses vocabulary
Output	Not always real words	Always meaningful
Example	"running" → "run" / "runn"	"running" → "run"
Accuracy	Low	High

Task 2:Build Advanced Preprocessing Function
Implement the following function:
def preprocess_text(text):
    pass


```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short words (≤2 except no, not)
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

Task 4: Token Analytics
For each processed sentence, compute:
Total number of tokens
Number of unique tokens
Average token length
Analysis Questions
Which sentence had the most noise?
Which sentence retained the most meaningful tokens after cleaning?


```python
def token_analytics(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg_len = token_analytics(tokens)

    print("\nSentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg_len)
```

Task 5: Frequency Analysis
Combine all tokens and use:
from collections import Counter

Identify:
Top 10 most frequent words
Top 5 least frequent words


```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

Task 6: Build Full Pipeline
def full_pipeline(text_list):
    pass

Expected Output
{
    "tokens": [...],
    "clean_sentences": [...]
}





```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Run pipeline
result = full_pipeline(test_sentences)
print(result)
```
