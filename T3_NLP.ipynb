```python
# NLP Assignment 3 – Chatbot using Transformers
# Name: ANUSHKA
# College: Your College

# Install libraries (run once)
!pip install transformers torch

# Import libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

print("Loading chatbot... please wait ⏳")

# Load model
model_name = "microsoft/DialoGPT-medium"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print("\nChatbot: Hello! I am your AI assistant. How can I help you today?\n")

chat_history_ids = None

# Chat loop
while True:
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a great day 😊")
        break

    # Encode input
    new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

    # Manage chat history
    if chat_history_ids is not None:
        bot_input_ids = torch.cat([chat_history_ids[:, -500:], new_input_ids], dim=-1)
    else:
        bot_input_ids = new_input_ids

    # Generate response
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=1000,
        min_length=20,
        do_sample=True,
        top_k=100,
        top_p=0.92,
        temperature=0.7,
        repetition_penalty=1.2,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode response
    response = tokenizer.decode(
        chat_history_ids[:, bot_input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    print("Chatbot:", response)
```
